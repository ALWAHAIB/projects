```python
import pandas as pd
import pandapower as pp
import pandapower.networks as nw

#Create an empty network 
net = pp.create_empty_network()
```

```python
# Create double busbar to connect  form Ext_Grid to our Grid system (380Kv)
pp.create_bus(net, name='Double BB 1', vn_kv=380, type='b')
pp.create_bus(net, name='Double BB 2', vn_kv=380, type='b')
```

```python
# Create nodes for EHV Feeders lets say we have 3 feeders conncted to the Double Bus so here we need to create 2*3 Nodes
for i in range(6): 
    pp.create_bus(net, name='GIS T%s' % i, vn_kv=380, type='n')
# Now we create 
for i in range(1, 3):
    pp.create_bus(net, name='GIS %s' % i, vn_kv=380, type='n')
# Single busbar
pp.create_bus(net, name='Single Busbar', vn_kv=110, type='b')
for i in range(1, 6):
    pp.create_bus(net, name='Bus SB %s' % i, vn_kv=110, type='n')
for i in range(1, 6):
    for j in [1, 2]:
        pp.create_bus(net, name='Bus SB T%s.%s' % (i, j), vn_kv=110, type='n')

# Remaining buses
for i in range(1, 5):
    pp.create_bus(net, name='Bus HV%s' % i, vn_kv=110, type='n')

# show bustable
net.bus
```
