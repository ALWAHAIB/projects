# Data visualisation task 

**Additional information**  
[Cars93 metadata](https://jse.amstat.org/v1n1/datasets.lock.html) 

```python
# Import libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('Cars93.csv')
```

### 1. Boxplot

```python
# Boxplot for revs per mile
manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
df_filtered = df[df['Manufacturer'].isin(manufacturers)] 
plt.figure()
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=df_filtered)
plt.title('Revs per Mile by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Revs per Mile')
plt.show()
plt.close()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer: The plot shows that Suzuki have the highest Revs/mile (only one car manufactured by Suzuki in the Data)**


### 2. Histogram

```python
# Histogram of MPG
plt.figure()
plt.hist(df['MPG.city'], alpha=0.5, label='City MPG')
plt.hist(df['MPG.highway'], alpha=0.5, label='Highway MPG')
plt.title('Distribution of MPG in City and Highway')
plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
plt.close()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer: The disrobution graph shows driving on the city is more efficient**

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure()
sns.lineplot(x=df['Wheelbase'], y=df['Turn.circle'],marker='o', linestyle='-')
plt.title('Relationship between Wheelbase and Turning Circle')
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
plt.show()
plt.close()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer: as the wheelbase get larger the turning circle get bigger, thus it have proportional relationship**  


### 4. Barplot

```python
# Barplot for mean horsepower

plt.figure()
sns.barplot(x=df['Type'], y=df['Horsepower'])
plt.title('Mean Horsepower by Car Type')
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.show()
plt.close()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer: As the data shows yes, larger cars tend to have more horsepower**
